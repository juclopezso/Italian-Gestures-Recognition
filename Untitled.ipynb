## HOG

```python
def hog_features(image_uri):
    ppc=14
    cpb=(2,2)
    img_size=(160,160)

    image = cv2.imread(image_uri)
    image = cv2.resize(image, img_size)
    #image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    fd,hog_image = hog(image, orientations=8, pixels_per_cell=(ppc,ppc),
                       cells_per_block=cpb,block_norm='L2',visualize=True)

    return fd, hog_image
```

```python
def hog_features_camera(image):
    ppc=14
    cpb=(2,2)
    img_size=(160,160)

    #image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    fd,hog_image = hog(image, orientations=8, pixels_per_cell=(ppc,ppc),
                       cells_per_block=cpb,block_norm='L2',visualize=True)

    return fd, hog_image
```

## Window sliding

```python
def sliding_window(image, stepSize, windowSize):
    # slide a window across the image
    for y in range(0, image.shape[0], stepSize):
        for x in range(0, image.shape[1], stepSize):
            # yield the current window
            yield (x, y, image[y:y + windowSize[1], x:x + windowSize[0]])
    

def pyramid(image):
    #image1 = cv2.resize(image, (600,600))
    image1 = cv2.resize(image, (400,400))
    image2 = cv2.resize(image, (270,270))
    
    return image1, image2#, image3
```

```python
video_capture = cv2.VideoCapture(0)
(winW, winH) = (160,160)

while True:
    # Capture frame-by-frame
    ret, frame = video_capture.read()
    
    # loop over the image pyramid
    for resized in pyramid(frame):
        # loop over the sliding window for each layer of the pyramid
        for (x, y, window) in sliding_window(resized, stepSize=50, windowSize=(winW, winH)):
            # if the window does not meet our desired window size, ignore it
            if window.shape[0] != winH or window.shape[1] != winW:
                continue
    
            h_f, h_i = hog_features_camera(window)
            pred = clf.predict(h_f.reshape(1, -1))[0]

            sentiment = ""

            if pred == 1:
                sentiment = "Malhumorado"
            if pred == 2:
                sentiment = "Disgustado"
            if pred == 3:
                sentiment = "Feliz/Satisfecho"
            if pred == 4:
                sentiment = "Desesperado"
            if pred == 5:
                sentiment = "Alegre"
            if pred == 6:
                sentiment = "Furioso"
            
            clone = resized.copy()
            cv2.putText(clone, "Clase: " + str(pred), (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
            cv2.putText(clone,  sentiment , (10, 80), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
            cv2.rectangle(clone, (x, y), (x + winW, y + winH), (0, 255, 0), 2)
            cv2.imshow("Window", clone)
            time.sleep(0.025)

            if cv2.waitKey(1) & 0xFF == ord('q'):
                break

# When everything is done, release the capture
video_capture.release()
cv2.destroyAllWindows()
```
