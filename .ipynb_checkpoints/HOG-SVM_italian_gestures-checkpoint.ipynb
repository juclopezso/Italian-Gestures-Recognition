## Data augmentation (not used)

```python
datagen = ImageDataGenerator(
        rotation_range=22,
        width_shift_range=0.2,
        height_shift_range=0.2,
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.05,
        horizontal_flip=True,
        fill_mode='nearest') #the strategy used for filling in newly created pixels

# the .flow() command below generates batches of randomly transformed images
# and saves the results to the `preview/` directory



for i in range(1,6):
    folder = str(i)
    for j in range(1,59):
        if j<10:
            image_route = "data/train/"+folder+"/archivo0000%d.png" %j
        else:
            image_route = "data/train/"+folder+"/archivo000%d.png" %j
        
        print(image_route)
        
        img = load_img(image_route)
        x = img_to_array(img)
        x = x.reshape((1,) + x.shape)
            
        k = 0
        for batch in datagen.flow(x, batch_size=1, save_to_dir="preview/train/"+folder, save_prefix='archivo', save_format='png'):
            k += 1
            if k > 20:
                break  # otherwise the generator would loop indefinitely
                
    
```

## HOG 

```python
def hog_features(image_uri):
    ppc=16
    cpb=(2,2)
    img_size=(160,160)

    image = cv2.imread(image_uri)
    image = cv2.resize(image, img_size)
    #image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    fd,hog_image = hog(image, orientations=8, pixels_per_cell=(ppc,ppc),
                       cells_per_block=cpb,block_norm='L2',visualize=True)

    return fd, hog_image
```

## HOG features for train dataset

```python
import glob

#train_data = {}
train_features = []
train_images = []
train_labels = []

train_size = 600
classes = 5

for i in range(1,classes+1):
    folder = str(i)
    for j in range(1,train_size+1):
        image_route = "data/train-videos/"+folder+"/frame%d.jpg" %j
            
        print(image_route)
        h_f, h_i = hog_features(image_route)
            
        train_features.append(h_f)
        train_images.append(h_i)
        train_labels.append(i)
```

## HOG features for validation dataset

```python
validation_features = []
validation_images = []
validation_labels = []
images = []

validation_size = 14


for i in range(1,6):
    folder = str(i)
    #train_data[i] = []
    for j in range(1,validation_size+1):
        h_f, h_i = hog_features("data/validation-videos/"+folder+"/archivo%d.jpg" %j)
        images.append(cv2.imread("data/validation-videos/"+folder+"/archivo%d.jpg" %j))
            
        validation_features.append(h_f)
        validation_images.append(h_i)
        validation_labels.append(i)
```

### Fit the dataset using SVM (Train)

```python
clf = svm.SVC()

#hog_features = np.array(hog_features)
#data_frame = np.hstack((hog_features,labels))
#np.random.shuffle(data_frame)

x_train, x_test = train_features[:], validation_features[:]
y_train, y_test = train_labels[:] , validation_labels[:]

clf.fit(x_train,y_train)
```

## Test the model

```python
y_pred = clf.predict(x_test)

print("Accuracy: "+ str(accuracy_score(y_test, y_pred)))
print('\n')
print(classification_report(y_test, y_pred))
```

```python
def hog_features_camera(image):
    ppc=16
    cpb=(2,2)
    img_size=(160,160)

    #image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    fd,hog_image = hog(image, orientations=8, pixels_per_cell=(ppc,ppc),
                       cells_per_block=cpb,block_norm='L2',visualize=True)

    return fd, hog_image
```

```python
video_capture = cv2.VideoCapture(0)

while True:
    # Capture frame-by-frame
    ret, frame = video_capture.read()
    
    image = cv2.resize(frame, (160,160))
    frame = cv2.resize(frame, (500,370))
    
    h_f, h_i = hog_features_camera(image)
    pred = clf.predict(h_f.reshape(1, -1))[0]
    
    sentiment = ""
    
    if pred == 1:
        sentiment = "Malhumorado"
    if pred == 2:
        sentiment = "Disgustado"
    if pred == 3:
        sentiment = "Feliz/Satisfecho"
    if pred == 4:
        sentiment = "Desesperado"
    if pred == 5:
        sentiment = "Alegre"

    cv2.putText(frame, "Clase: " + str(pred), (10, 50), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
    cv2.putText(frame,  sentiment , (10, 80), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)

    # Display the resulting frame
    cv2.imshow('Video', frame)
    

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# When everything is done, release the capture
video_capture.release()
cv2.destroyAllWindows()
```
